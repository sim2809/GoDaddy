```python
df_train = pd.concat([df_train, pd.get_dummies(df_train["county"])],  axis=1)
```

```python
df_train = pd.concat([df_train, pd.get_dummies(df_train["state"])],  axis=1)
```

```python
df_train.drop(["state","county"], axis=1, inplace=True)
```

```python
df_train.drop(["row_id","cfips"], axis=1, inplace=True)
```

```python
target = df_train["microbusiness_density"]
df_train.drop(["microbusiness_density"], axis=1, inplace=True)
```

```python
import datetime as dt
df_train['Date'] = pd.to_datetime(df_train['first_day_of_month'])
df_train['Date']=df_train['Date'].map(dt.datetime.toordinal)
```

```python
df_train.order_by("Data")
```

```python
df_train.drop(["first_day_of_month", "active"], axis=1, inplace=True)
```

```python
df_train.sort_values(by='Date', inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_train, target, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=4)
neigh.fit(X_train, y_train)
```

```python
y_pred = neigh.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
